```python
import yfinance as yf
import pandas as pd

# 1. Download Apple stock data as an example
ticker = 'AAPL'
data = yf.download(ticker, start='2020-01-01', end='2024-01-01')

# 2. View the first few rows (you should see Open, High, Low, Close)
data.head()
```

```python
# Select only the 'Close' prices for AAPL
df = data['Close'].copy()

# Rename the column to 'Close' so it's clean
df.columns = ['Close']

# View the clean version
df.head()
```

```python
# 1. Simple Moving Average (SMA) - 20 days
df['SMA_20'] = df['Close'].rolling(window=20).mean()

# 2. Exponential Moving Average (EMA) - 20 days
df['EMA_20'] = df['Close'].ewm(span=20, adjust=False).mean()

# 3. Daily Returns (%)
df['Daily_Return'] = df['Close'].pct_change() * 100

# Look at the results
df.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Apple Close Price', alpha=0.5)
plt.plot(df['SMA_20'], label='20-Day SMA', color='orange')
plt.plot(df['EMA_20'], label='20-Day EMA', color='green')
plt.title('Apple Price vs. Moving Averages')
plt.legend()
plt.show()
```

```python
# --- MACD Calculation ---
exp1 = df['Close'].ewm(span=12, adjust=False).mean()
exp2 = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = exp1 - exp2
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()

# --- RSI Calculation (14-day) ---
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))

print("RSI and MACD calculated!")
df.tail()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['RSI'], color='purple', label='RSI')
plt.axhline(70, linestyle='--', color='red', alpha=0.5) # Overbought line
plt.axhline(30, linestyle='--', color='green', alpha=0.5) # Oversold line
plt.title('RSI (Momentum Indicator)')
plt.legend()
plt.show()
```

```python
# Save your final data with all the indicators
df.to_csv('../data/processed_stock_data_with_indicators.csv')
print("CSV Saved Successfully!")
```
