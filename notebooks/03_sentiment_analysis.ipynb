```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# 1. Setup the sentiment tool
nltk.download('vader_lexicon')
vader = SentimentIntensityAnalyzer()

# 2. Load your news data
# (Based on yesterday, your file is in data/raw/)
df_news = pd.read_csv('../data/raw/raw_analyst_ratings.csv')

# 3. Calculate Sentiment (This gives a score from -1 to 1)
def get_score(text):
    return vader.polarity_scores(str(text))['compound']

# Let's test it on the first 1000 rows first to be fast
df_news_sample = df_news.head(1000).copy()
df_news_sample['sentiment_score'] = df_news_sample['headline'].apply(get_score)

# 4. View results
df_news_sample[['headline', 'sentiment_score']].head()
```

```python
# 1. Convert the date column to actual dates
df_news['date'] = pd.to_datetime(df_news['date'], errors='coerce')

# 2. Run sentiment on the WHOLE dataset (this might take a few minutes)
# We use a progress bar logic or just wait for the [*] to turn into a number
df_news['sentiment_score'] = df_news['headline'].apply(get_score)

# 3. Group by Date and Stock to get the average daily sentiment
daily_sentiment = df_news.groupby(['stock', df_news['date'].dt.date])['sentiment_score'].mean().reset_index()

# 4. View your new daily sentiment table
daily_sentiment.head()
```
