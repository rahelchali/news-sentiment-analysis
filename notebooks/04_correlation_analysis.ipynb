```python
pythonimport pandas as pd
import matplotlib.pyplot as plt

# 1. Load your processed data
prices = pd.read_csv('../data/processed_stock_data_with_indicators.csv')
sentiment = pd.read_csv('../data/daily_sentiment_summary.csv')

# 2. Match the date formats
prices['Date'] = pd.to_datetime(prices['Date']).dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

# 3. Merge Price and Sentiment (Focusing on AAPL)
df_final = pd.merge(prices, sentiment[sentiment['stock'] == 'AAPL'], left_on='Date', right_on='date')

# 4. Calculate the Correlation Score
# Measures relationship from -1 (opposite) to +1 (perfect match)
score = df_final['Close'].corr(df_final['sentiment_score'])
print(f"Final Project Result: The correlation score is {score:.4f}")

# 5. Visualize the Relationship
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(df_final['Date'], df_final['Close'], color='blue', label='Stock Price')
ax1.set_ylabel('Price ($)', color='blue')

ax2 = ax1.twinx()
ax2.bar(df_final['Date'], df_final['sentiment_score'], color='red', alpha=0.3, label='Sentiment')
ax2.set_ylabel('Sentiment Score', color='red')

plt.title('Final Analysis: Price vs. News Sentiment')
plt.show()
```
